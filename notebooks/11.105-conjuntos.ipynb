# Sets
Un set es una colección de objetos (únicos e inmutables) no ordenada que soportan operaciones de teoría de conjuntos.

```python
set([1, 2, 3, 4])
```

```python
{1, 2, 3, 4}
```

```python
set("spam")  # Añade los caracteres del string como objetos
# Los elementos no tienen orden
```

```python
S = {"s", "p", "a", "m"}
S.add("alot")  # 'alot' se añade como un único string
S
```

## Operaciones sobre conjuntos
`sets` soportan operaciones de conjuntos/
No pueden realizarse operaciones en sercuencias como en strings, listas o tuplas.

```python
S1 = {1, 2, 3, 4}  # Interseccion
S1 & {1, 3}
```

```python
{1, 5, 3, 6} | S1  # Union
```

```python
S1 - {1, 3, 4}  # Diferencia
```

```python
S1 > {1, 3}  # Superset
```

`{}` es un diccionario en todas las versiones de Python.
Los conjuntos vacíos deben ser creados con el built-in `set`, y se printan del mismo modo:

```python
S1 - {1, 2, 3, 4}  # Los conjuntos vacíos se printan con el built-in set
```

```python
type({})  # {} es un diccionario vacio
```

```python
S = set()  # Inicializamos un conjunto vacio
S.add(1.23)
S
```

## Vistas diccionario y sets
La vista objeto por el método `keys`es como un `set` y soporta operaciones de conjuntos

```python
D = dict(a=1, b=2, c=3)  # D = {'b': 2, 'c': 3, 'a': 1}
K = D.keys()
V = D.values()
del D["b"]
D
```

```python
K, V
```

```python
K | {"x": 4}  # La vista `keys` es como un `set`
```

```python
V & {"x": 4}  # La vista values no es como un `set`.
```

```python
V & {"x": 4}.values()  # La vista values no es como un `set`.
```
