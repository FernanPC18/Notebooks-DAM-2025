## Length
Vemos la longitud del string:

```python
len(S)
```

## Posiciones en Strings
En Python, los índices empiezan por 0, y los negativos cuentan hacia atrás empezando por el final (-1 es el último ítem).

```python
S[0] # 1er Item
```

```python
S[1] # 2o item
```

```python
S[8] # Saliendo de rango
```

```python
S[-1] # Ultimo item
```

```python
S[-2] # Penultimo item
```

## Backus-Naur

```python
S[len(S) -1] # Lo mismo que hacer S[-1]
```

## Slicing

```python
S[1:3] # Del 1 al 2 (3 no incluido)
```

```python
S[1:] # De 1 hasta el final
```

```python
S # S no ha cambiado
```

```python
S[0:3] # Todo excepto el último
```

```python
S[:3] # Lo mismo que S[0:3]
```

```python
S[:-1] # Todo excepto el último, pero más simple
```

```python
S[:] # Hace una copia que apunta al mismo punto en memoria
```

```python
S + 'eggs'
```

```python
S # no cambia
```

```python
S * 6
```

## Polimorfismo
En python el `+` tiene varias funciones, como sumar y concatenar. Qué se hace con ellos depende de los dos valores utilizados para operar con él, que serán comprobados para ver si tienen una operación compatible. A esto se le conoce como polimorfismo.

## Podemos usar expresiones para crear objetos nuevos

```python
S = 'z' + S[1:]
S
```

## Strings y listas

```python
E = 'egss'
L = list(E)
print(L)
L[0] = 'z'
L
```

```python
''.join(L)
# L es ['z', 'g', 's', 's']
```

## Métodos de tipo específico
### Find

```python
# Encuentra el offset de un substring en S
S = 'SpamEggsSpam'
S.find('pa')
```

```python
S.find('pa', 3)
```

```python
S.find(S)
```

```python
S.find('z')
```

### Replace

```python
S.replace('Eggs', 'Bacon') # S Es inmutable, asi que no cambia su valor
```

### Mayúsculas

```python
S.upper() # S es inmutable asi que no cambia su valor
```

```python
print("S is alpha", S.isalpha())
print("S is digit", S.isdigit())
```

### Rstrip
Elimina espacios en blanco a la derecha

```python
line = 'aaa\t,bbb\t,\nccccc,\tdd\n'
print(line)
```

```python
line.rstrip()
```

```python
line.lstrip()
```

```python
line.rstrip().split(',')
```

### Formatting
Strings soportan un método de reemplazo de texto conocido como **formatting**

```python
'%s, eggs, and %s' % ('spam', 'SPAM!')
```

```python
'{}, eggs, and {}'.format('spam', 'SPAM!')
```

```python
S = 'SpamEggsSpam'
f'{S}'
```

```python
'{0}'.format(S)
```

```python
ord('s')
```
